```python
import pandas as pd
```

```python
data = pd.read_csv("Live.csv")
```

```python
data
```

```python
ex_missing_values = data.isnull().sum()
print(ex_missing_values)
```

```python
data.drop(['Column1', 'Column2', 'Column3', 'Column4'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
data["status_type"] = labelencoder.fit_transform(data["status_type"].values)
```

```python
data
```

```python
data.drop("status_published", axis=1,inplace=True)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(data)
```

```python
labels = model.predict(data)
```

```python
import numpy as np
```

```python
np.unique(labels,return_counts = True)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4)
model.fit(data)
labels = model.predict(data)
import numpy as np
np.unique(labels,return_counts=True)
```

```python
data["labels"] = labels
data["labels"].value_counts()
```
