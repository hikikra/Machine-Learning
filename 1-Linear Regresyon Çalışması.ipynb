```python
import pandas as pd
```

```python
data =pd.read_csv("Hitters.csv")
```

```python
data.shape
```

```python
ex_missing_values = data.isnull().sum()
print(ex_missing_values)
```

```python
data = data.fillna(data.mean())
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
data["NewLeague"] = labelencoder.fit_transform(data["NewLeague"].values)
```

```python
data["League"] = labelencoder.fit_transform(data["League"].values)
```

```python
data["Division"] = labelencoder.fit_transform(data["Division"].values)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(data,test_size =0.3)
```

```python
x_train = train.drop("Salary",axis = 1)
y_train = train.loc[: , "Salary"]
```

```python
x_test = train.drop("Salary",axis = 1)
y_test = train.loc[: , "Salary"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
```

```python
comparison = pd.DataFrame({"Actual Valuse": y_test,"Predictions":predictions})
```

```python
comparison.head()
```

```python
comparison.tail()
```

```python
from sklearn.metrics import mean_squared_error
from numpy import sqrt
```

```python
rmse = sqrt(mean_squared_error(y_test,predictions))
rmse
```
